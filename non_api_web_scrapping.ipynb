```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
webpage=requests.get('https://www.ambitionbox.com/list-of-companies?sort_by=popularity&designation=data-scientist&page=1',headers=headers).text
```

```python
soup=BeautifulSoup(webpage,'lxml')
```

```python
# To find review count
soup.find_all('a',class_ = 'review-count')
```

```python
len(soup.find_all('a',class_ = 'review-count'))
```

```python
company = soup.find_all('div', class_='company-content-wrapper')
```

```python
company
```

```python
len(company)
```

```python

name=[]
rating=[]
reviews=[]
ctype=[]
hq=[]
how_old=[]
no_of_employee=[]

for i in company:

  name.append(i.find('h2').text.strip())
  rating.append(i.find('p',class_='rating').text.strip())
  reviews.append(i.find('a', class_='review-count').text.strip())
  try:
    ctype.append(i.find_all('p',class_='infoEntity')[0].text.strip())
  except:
    ctype.append(np.nan)
  try:
    hq.append(i.find_all('p',class_='infoEntity')[1].text.strip())
  except:
    hq.append(np.nan)
  try:
    how_old.append(i.find_all('p',class_='infoEntity')[2].text.strip())
  except:
    how_old.append(np.nan)
  try:
    no_of_employee.append(i.find_all('p',class_='infoEntity')[3].text.strip())
  except:
    no_of_employee.append(np.nan)


```

```python
len(name),len(rating),len(reviews),len(ctype),len(hq),len(how_old),len(no_of_employee)
```

```python
df=pd.DataFrame({'name':name,
   'rating':rating,
   'reviews':reviews,
   'company_type':ctype,
   'Head_Quarters':hq,
   'Company_Age':how_old,
   'No_of_Employee':no_of_employee,
   })
```
